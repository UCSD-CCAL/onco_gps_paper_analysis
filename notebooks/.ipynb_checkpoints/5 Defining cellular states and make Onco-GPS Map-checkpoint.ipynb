```python
mutation__gene_x_ccle_cellline = read_gct(
    '../data/mutation__gene_x_ccle_cellline.gct')

kras_mutants = mutation__gene_x_ccle_cellline.columns[
    mutation__gene_x_ccle_cellline.loc['KRAS_MUT'].astype(bool)]
```

```python
h_matrix = read_gct('../results/nmf_k9_h.gct')

kras_mutant_h_matrix = h_matrix[h_matrix.columns & kras_mutants]
```

```python
plot_heatmap(
    kras_mutant_h_matrix,
    normalization_method='-0-',
    normalization_axis=1,
    cluster=True,
    xlabel='Cellline',
    ylabel='KRAS Component',
    xticklabels=False)
```

```python
kras_h_matrix = kras_mutant_h_matrix.loc[['C7', 'C2', 'C1']]
```

```python
# '-0-' normalize each row
a = normalize_2d_array(kras_h_matrix.values, '-0-', axis=1)
# Clip values 3 standard deviation away from each row
a = a.clip(min=-3, max=3)
# '0-1' normalize each row
a = normalize_2d_array(a, '0-1', axis=1)

kras_h_matrix = pd.DataFrame(
    a, index=kras_h_matrix.index, columns=kras_h_matrix.columns)

ds, hcs, cs, cccs = define_states(
    kras_h_matrix,
    range(2, 8),
    '../results',
    n_clusterings=50,
    random_seed=12765)
```

```python
cs = pd.read_table('../results/hcc.txt', index_col=0)
cs
```

```python
kras_sample_labels = cs.loc['K4']
kras_sample_labels.name = 'State'
```

```python
plot_heatmap(
    kras_h_matrix,
    data_type='continuous',
    normalization_method='-0-',
    normalization_axis=1,
    column_annotation=kras_sample_labels,
    annotation_colors=[PURPLE, BLUE, RED, GREEN],
    figure_size=(16, 16),
    title='Clustered KRAS H-Matrix',
    xlabel='Sample',
    ylabel='Component',
    xticklabels=False)
```

```python
gps_map = GPSMap(
    kras_h_matrix,
    pull_power=1.2,
    node_x_dimension=pd.DataFrame(
        [
            [0.5, np.sqrt(3) / 2],
            [1, 0],
            [0, 0],
        ], index=kras_h_matrix.index))
```

```python
gps_map.set_sample_phenotypes(
    kras_sample_labels,
    phenotype_type='categorical',
#     phenotype_type='continuous',
    n_grids=256,
    bandwidth_factor=2.8,
        phenotype_color_map=mpl.colors.ListedColormap([PURPLE, BLUE, RED, GREEN]),
#     phenotype_color_map=mpl.cm.bwr,
    phenotype_to_str={
        0: 'State 1',
        1: 'State 2',
        2: 'State 3',
        3: 'State 4',
    },
)
```
